```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() using object oriented method
fig , ax=plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() another way of writing code for above code
fig = plt.figure()
ax = fig.add_subplot()
ax.plot (x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis , y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel('Fare($)')

## Set the y limit between 0 and 45.
plt.ylim(0,45)


# Create a title
plt.title('Pyber Fare by Month')

# Add a grid
plt.grid()

# Add the legend
plt.legend()

#change the figure size to 20 pixels wide by 10 pixels high
plt.figure(figsize=(20, 102))
```

```python
# Create the plot.
fig , ax=plt.subplots()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')

# Add a label 
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

# Add a title.
ax.set_title("Pyber Fare by Month")

# Sets the min and max range of the y-axis
ax.set_ylim(0, 45)

# Add a grid to the chart.
ax.grid()

#Add a legend.
ax.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c='red', label='Chicago')

plt.xlabel('Fare($)')
plt.ylabel('Date')

# Create a title.
plt.title("PyBer Fare by Month")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create Scatter plot
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# make size larger
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
# we can write in this way as well
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# scatter plot using object oriented interface
fig, ax=plt.subplots()
ax.scatter(x_axis,  y_axis)
```

```python
# bubble chart using object oriented
fig, ax=plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', s=[i*5 for i in y_axis], alpha=0.80, edgecolors='black', linewidths=2, 
           label='Boston')
ax.set_title("Pyber Fare by Month")
ax.legend()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.invert_yaxis()
ax.set_xlim(0, 50)
```

```python
# Create pie chart
plt.pie(y_axis , labels= x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels= x_axis, autopct= '%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", 
          "orange", "gold", "indianred", "tomato", "mistyrose"]

explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# pie chart using object oriented
fig, ax=plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", 
          "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, shadow='true', counterclock= False,
      startangle= 97, colors=colors)

plt.show
```
